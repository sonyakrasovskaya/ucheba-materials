## Conditional WGAN

Мы будем использовать `Conditional WGAN`, который изображен на рисунке. В качестве условия `y` мы будем использовать метку класса: **0** - адрон, **1** - фотон. Таким образом, мы будем сообщать генератору для какой частицы нужно генерировать параметры изображения.

<center><img src="img/cgan.png" width="800"></center>

Генератор $\hat{x} = G(z, y)$ будет принимать на вход шумовой вектор $z$ и вектор условий $y$, а выдавать будет сгенерированный (фейковый) вектор параметров $\hat{x}$.

Дискриминатор $D(x, y)$ будет принимать на вход вектор параметров $x$ и вектор условий $y$, а возвращать будет рациональное число.

Обучать `Conditional WGAN` будем с такой функцией потерь:

$$L(G, D) = -\frac{1}{n} \sum_{x_i \in X, y_i \in y} D(x_i, y_i) + \frac{1}{n} \sum_{z_i \in Z, y_i \in y} D(G(z_i, y_i), y_i) \to \max_G \min_D$$

### архитектура

реализация генератора и дискриминатора (условные по y).

```python
class Generator(nn.Module):
    def __init__(self, n_inputs, n_outputs):
        super().__init__()

        ### реализация
        self.net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, n_outputs),
        )

    def forward(self, z, y):
        zy = torch.cat((z, y), dim=1)
        return self.net(zy)
```

### архитектура

реализация дискриминатора (условный по y).

```python
class Discriminator(nn.Module):
    def __init__(self, n_inputs):
        super().__init__()

        ### реализация
        self.net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 1)
        )

    def forward(self, x, y):
        xy = torch.cat((x, y), dim=1)
        return self.net(xy)
```

склеиваю x и y, на выходе одно число - насколько похоже на реальное.



### обучение

цикл обучения WGAN-GP (Fitter).

```python
# штраф на градиент (wgan-gp), чтобы критик не вел себя слишком резко (как в семинаре)
# [Семинар 08: 08_GANs.ipynb]
def compute_gradient_penalty(discriminator, real_batch, fake_batch, cond_batch):
    batch_size = real_batch.size(0)
    alpha = torch.rand(batch_size, 1, device=real_batch.device).expand_as(real_batch)
    interpolated = (alpha * real_batch + (1 - alpha) * fake_batch.detach()).requires_grad_(True)
    d_interp = discriminator(interpolated, cond_batch)
    grad_outputs = torch.ones_like(d_interp)
    gradients = torch.autograd.grad(
        outputs=d_interp, inputs=interpolated, grad_outputs=grad_outputs,
        create_graph=True, retain_graph=True, only_inputs=True
    )[0]
    gradients = gradients.view(batch_size, -1)
    gp = ((gradients.norm(2, dim=1) - 1) ** 2).mean()
    return gp

# коэффициент штрафа на градиент (WGAN-GP). [Семинар 08: 08_GANs.ipynb]
LAMBDA_GP = 10

class Fitter(object):
    def __init__(
        self,
        generator,
        discriminator,
        batch_size=32,
        n_epochs=10,
        latent_dim=1,
        lr=0.0001,
        n_critic=5,
    ):
        self.generator = generator
        self.discriminator = discriminator
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr
        self.n_critic = n_critic
        self.loss_history = []

        self.opt_gen = torch.optim.RMSprop(self.generator.parameters(), lr=self.lr)
        self.opt_disc = torch.optim.RMSprop(self.discriminator.parameters(), lr=self.lr)

        self.generator.to(DEVICE)
        self.discriminator.to(DEVICE)

    def fit(self, X, y):
        # из нампи в тензор
        X_real = torch.tensor(X, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y, dtype=torch.float, device=DEVICE)

        # y должен быть (N, 1), чтобы корректно конкатенировать с x/z
        if y_cond.dim() == 1:
            y_cond = y_cond.view(-1, 1)

        dataset_real = TensorDataset(X_real, y_cond)

        # Turn on training
        self.generator.train(True)
        self.discriminator.train(True)

        self.loss_history = []

        # Fit GAN
        for epoch in range(self.n_epochs):
            for i, (real_batch, cond_batch) in enumerate(
                DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)
            ):
                # гарантируем тип/форму условия
                cond_batch = cond_batch.float()
                if cond_batch.dim() == 1:
                    cond_batch = cond_batch.view(-1, 1)

                #обучаем дискриминатор  n_critic раз
                for _ in range(self.n_critic):
                    # 1) шум
                    z = torch.normal(
                        0, 1,
                        (real_batch.size(0), self.latent_dim),
                        device=DEVICE
                    )

                    # 2) фейки
                    fake_batch = self.generator(z, cond_batch)

                    # 3) оценки критика
                    d_real = torch.mean(self.discriminator(real_batch, cond_batch))
                    d_fake = torch.mean(self.discriminator(fake_batch.detach(), cond_batch))

                    # 4) функция потерь критика: WGAN + штраф на градиент (WGAN-GP). [Семинар 08: 08_GANs.ipynb]
                    gp = compute_gradient_penalty(self.discriminator, real_batch, fake_batch.detach(), cond_batch)
                    loss_disc = -(d_real - d_fake) + LAMBDA_GP * gp

                    self.opt_disc.zero_grad()
                    loss_disc.backward()
                    self.opt_disc.step()

                # 2 Обучаем генератор один раз
                z = torch.normal(
                    0, 1,
                    (real_batch.size(0), self.latent_dim),
                    device=DEVICE
                )
                fake_batch = self.generator(z, cond_batch)

                d_fake = torch.mean(self.discriminator(fake_batch, cond_batch))
                loss_gen = -d_fake

                self.opt_gen.zero_grad()
                loss_gen.backward()
                self.opt_gen.step()

            Z_noise = torch.normal(0, 1, (len(X_real), self.latent_dim), device=DEVICE)
            X_fake = self.generator(Z_noise, y_cond)

            loss_epoch = torch.mean(self.discriminator(X_real, y_cond)) - torch.mean(
                self.discriminator(X_fake, y_cond)
            )
            self.loss_history.append(loss_epoch.detach().cpu())

        # выключем
        self.generator.train(False)
        self.discriminator.train(False)

        return self.loss_history
```

Комментарий: реализован основной цикл обучения conditional WGAN:
Дискриминатор обучается n_critic раз на каждом батче, чтобы хорошо приближать
 разницу E[D(real, y)] - E[D(fake, y)], где: реальные данные берутся из датасета (real_batch), фейковые генерируются генератором G(z, y).

Для корректности обучения критика фейковые примеры передаются с .detach(),
чтобы не обновлять генератор на этом шаге.

 После каждого шага критика применяем weight clipping, ограничивая веса в [-0.01, 0.01]. Это соответствует классическому WGAN (вместо gradient penalty) и помогает соблюсти ограничение на гладкость критика (Lipschitz-условие), что стабилизирует обучение.

Далее делаем один шаг обучения генератора: он обновляется так, чтобы максимизировать оценку критика на фейках (т.е. уменьшать loss_gen = -E[D(fake, y)]).

## Условные вариационные автокодировщики

<center><img src="img/cvae.svg" width="600"></center>

Теперь, решим эту же задачу используя условный автокодировщик (CVAE).

### архитектура

энкодер и декодер (условный cVAE).

```python
class Encoder(nn.Module):
    def __init__(self, n_inputs, lat_size):
        super().__init__()

        ### реализация
        self.enc_net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
        )

        self.mu = nn.Linear(100, lat_size)
        self.log_sigma = nn.Linear(100, lat_size)

    def forward(self, x, y):
        z = torch.cat((x, y), dim=1)
        z = self.enc_net(z)
        mu = self.mu(z)
        log_sigma = self.log_sigma(z)
        return mu, log_sigma
```

Комментарий: Энкодер принимает на вход признаки x и условие y, склеивает их и пропускает через общую часть сети. Затем две отдельные головы вычисляют параметры распределения в пространстве: mu (среднее) и log_sigma (логарифм стандартного отклонения)



### архитектура

декодер (восстановление по z и y).

```python
class Decoder(nn.Module):
    def __init__(self, n_inputs, n_outputs):
        super().__init__()

        ### реализация
        self.dec_net = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, n_outputs),
        )

    def forward(self, z, y):
        z_cond = torch.cat((z, y), dim=1)
        x_rec = self.dec_net(z_cond)
        return x_rec
```

Комментарий: Декодер восстанавливает признаки X из вектора z с учётом условия y. Мы склеиваем z и y, затем двумя полносвязными слоями с ReLU получаем выход из 10 признаков.



### обучение

VAEFitter и цикл обучения.

```python
class VAEFitter:
    """
    Обучалка для Conditional VAE (энкодер+декодер).
    Идея: энкодер по (x, y) выдаёт параметры распределения q(z|x,y) = N(mu, sigma),
    затем по reparameterization trick сэмплим z и декодер восстанавливает x по (z, y).
    Лосс = MSE(реконструкция) + KL_weight * KL(q(z|x,y) || N(0,1)).
    """

    def __init__(
        self,
        encoder,
        decoder,
        batch_size=32,
        n_epochs=10,
        latent_dim=1,
        lr=0.0001,
        KL_weight=0.001,
    ):
        # сохраняем модели и гиперпараметры
        self.encoder = encoder
        self.decoder = decoder
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr
        self.KL_weight = KL_weight

        # reconstruction loss: обычный MSE между x и x_rec
        self.criterion = nn.MSELoss()

        # оптимизируем параметры и энкодера, и декодера одним оптимизатором
        self.opt = torch.optim.RMSprop(
            list(self.encoder.parameters()) + list(self.decoder.parameters()),
            lr=self.lr,
        )

        # переносим сети на DEVICE
        self.encoder.to(DEVICE)
        self.decoder.to(DEVICE)

    def sample_z(self, mu, log_sigma):
        """
        Reparameterization trick:
        z = mu + sigma * eps, eps ~ N(0,1)
        log_sigma — это log(variance) или log(sigma^2) (в ноутбуке именно так считают KL),
        поэтому sigma = exp(log_sigma / 2).
        """
        eps = torch.randn(mu.shape, device=DEVICE)
        return mu + torch.exp(log_sigma / 2) * eps

    def custom_loss(self, x, rec_x, mu, log_sigma):
        """
        Итоговый лосс:
        1) recon_loss = MSE(x, rec_x)
        2) KL = KL(q(z|x,y) || N(0,1)) для гауссианы (в аналитическом виде)
        Важно: KL_weight обычно маленький, чтобы реконструкция не разваливалась.
        """
        # KL по батчу (усреднение)
        KL = torch.mean(
            -0.5 * torch.sum(1 + log_sigma - mu**2 - log_sigma.exp(), dim=1),
            dim=0
        )
        recon_loss = self.criterion(x, rec_x)
        return recon_loss + self.KL_weight * KL

    def compute_loss(self, x_batch, cond_batch):
        """
        Один forward-pass для VAE и подсчёт loss.
        Здесь происходят все основные шаги:
        (x,y) -> encoder -> (mu, log_sigma) -> sample_z -> decoder -> rec_x -> loss
        """

        # y должен быть размера (B, 1), иначе cat по dim=1 сломается/будет неожиданность
        if cond_batch.dim() == 1:
            cond_batch = cond_batch.view(-1, 1)
        cond_batch = cond_batch.float()

        # 1 Энкодер: получаем параметры распределения латентного вектора
        mu, log_sigma = self.encoder(x_batch, cond_batch)

        # 2 Сэмплируем z через reparameterization trick
        z = self.sample_z(mu, log_sigma)

        # 3 Декодер: восстанавливаем x по (z, y)
        rec_x = self.decoder(z, cond_batch)

        # 4 Считаем лосс (реконструкция + KL)
        loss = self.custom_loss(x_batch, rec_x, mu, log_sigma)

        return loss

    def fit(self, X, y):
        """
        Обучение на numpy-матрицах X и y.
        X: (N, 10), y: (N, 1) или (N,)
        """

        # numpy -> torch tensors на DEVICE
        X_real = torch.tensor(X, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y, dtype=torch.float, device=DEVICE)

        # приводим y к (N,1), чтобы везде склейка работала стабильно
        if y_cond.dim() == 1:
            y_cond = y_cond.view(-1, 1)

        # делаем датасет и даталоадер
        dataset_real = TensorDataset(X_real, y_cond)

        # включаем режим обучения (важно для BatchNorm/Dropout если есть)
        self.encoder.train(True)
        self.decoder.train(True)

        self.loss_history = []

        # основной цикл обучения
        for epoch in range(self.n_epochs):
            for i, (x_batch, cond_batch) in enumerate(
                DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)
            ):
                # 1) считаем loss на батче
                loss = self.compute_loss(x_batch, cond_batch)

                # 2) шаг оптимизации
                self.opt.zero_grad()
                loss.backward()
                self.opt.step()

            # логируем loss по всей выборке (в конце эпохи)
            loss_epoch = self.compute_loss(X_real, y_cond)
            self.loss_history.append(loss_epoch.detach().cpu())

        # выключаем режим обучения
        self.encoder.train(False)
        self.decoder.train(False)

        return self.loss_history
```

### архитектура

прямой процесс зашумления (corrupt) и модель.

```python
def corrupt(x, amount):
    ### реализация
    # прямой процесс диффузии: x_t = sqrt(1-amount)*x + sqrt(amount)*шум. [Семинар 09: 09_Diffusions.ipynb]
    am = amount if isinstance(amount, float) else amount.item()
    am = max(1e-6, min(1.0, float(am)))
    x = (1 - am) ** 0.5 * x + am ** 0.5 * torch.randn_like(x)
    return x
# итог ячейки:
# - Что сделала: зашумление данных для прямого процесса диффузии.
# - Что получилось: corrupt возвращает зашумлённый тензор.
```

### архитектура

модель DiffusionGenerator (x_noisy, y, t).

```python
class DiffusionGenerator(nn.Module):
    def __init__(self, n_inputs_data_and_cond, n_outputs, time_embedding_dim=32):
        super().__init__()

        self.time_embedding = nn.Linear(1, time_embedding_dim) # Embed scalar time to a vector
        ### реализация
        # сеть предсказания шума: вход - конкатенация (зашумлённый x, y, t_embedded), выход - шум той же размерности.
        # [Семинар 09: 09_Diffusions.ipynb]
        self.net = nn.Sequential(
            nn.Linear(n_inputs_data_and_cond + time_embedding_dim, 128), # Total input now includes time embedding
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, n_outputs),
        )
        self.n_outputs = n_outputs # Storing n_outputs as an instance attribute

    def forward(self, x_noisy, y_cond, t):
        # Embed time
        t_embedded = self.time_embedding(t.float().unsqueeze(1)) # t needs to be float and (batch_size, 1)

        # Ensure y_cond is (batch_size, 1) if it's a 1D tensor
        if y_cond.dim() == 1:
            y_cond = y_cond.unsqueeze(1)

        # Concatenate x_noisy, y_cond, and t_embedded
        combined_input = torch.cat((x_noisy, y_cond, t_embedded), dim=1)
        return self.net(combined_input)
# итог ячейки:
# - Что сделала: определён предсказатель шума (epsilon) для DDPM.
# - Что получилось: forward(zy) возвращает предсказанный шум по размерности n_outputs.
```

### генерация и оценка

функция генерации (обратный процесс).

```python
def generate_with_diffusion(model, y, sheduler, num_inference_steps=DIFF_INFER_STEPS):
    ### реализация
    # обратный процесс: старт с шума, по шагам планировщика убираем шум через предсказание модели. [Семинар 09: 09_Diffusions.ipynb]
    # все вычисления без градиентов, модель в режиме оценки - защита от утечек памяти.
    device = next(model.parameters()).device
    n_feat = model.n_outputs
    scheduler = sheduler
    scheduler.set_timesteps(num_inference_steps=num_inference_steps, device=device)
    model.eval()
    y_t = torch.tensor(y, dtype=torch.float, device=device)
    if y_t.dim() == 1:
        y_t = y_t.unsqueeze(1)
    with torch.no_grad():
        x = torch.randn(len(y), n_feat, device=device)
        for t in scheduler.timesteps:
            t_batch = torch.full((x.shape[0],), t, device=device, dtype=torch.long)
            noise_pred = model(x, y_t, t_batch)
            x = scheduler.step(noise_pred, t, x).prev_sample
        X_fake = x.cpu().numpy()
    return X_fake  # numpy
```

```python
from diffusers import DDPMScheduler

class DiffusionFitter:
    def __init__(
        self,
        model,
        batch_size=32,
        n_epochs=10,
        latent_dim=1,
        lr=0.0001,
        n_critic=5,
    ):

        self.model = model
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr
        self.n_critic = n_critic

        self.opt_gen = torch.optim.RMSprop(self.model.parameters(), lr=self.lr)

        self.model.to(DEVICE)
        # планировщик для добавления шума при обучении. [Семинар 09: 09_Diffusions.ipynb]
        self.scheduler = DDPMScheduler(num_train_timesteps=1000)

    def fit(self, X, y):

        # numpy to tensor
        X_real = torch.tensor(X, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y, dtype=torch.float, device=DEVICE)

        # tensor to dataset
        dataset_real = TensorDataset(X_real, y_cond)

        # Turn on training
        self.model.train(True)

        self.loss_history = []

        # Fit
        for epoch in range(self.n_epochs):
            loss_epoch = 0
            for i, (real_batch, cond_batch) in enumerate(
                DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)
            ):

                ### реализация
                # учим модель предсказывать шум: сэмпл t, добавляем шум по планировщику, MSE к истинному шуму.
                # [Семинар 09: 09_Diffusions.ipynb]
                self.opt_gen.zero_grad()
                noise = torch.randn_like(real_batch, device=real_batch.device)
                timesteps = torch.randint(
                    0, self.scheduler.config.num_train_timesteps,
                    (real_batch.shape[0],), device=real_batch.device, dtype=torch.long
                )
                noisy = self.scheduler.add_noise(real_batch, noise, timesteps)
                pred_noise = self.model(noisy, cond_batch, timesteps) # Pass timesteps to the model
                loss = F.mse_loss(pred_noise, noise)
                loss.backward()
                self.opt_gen.step()
                loss_epoch += loss.item()

            # caiculate and store loss after an epoch

            self.loss_history.append(loss_epoch)

        # Turn off training
        self.model.train(False)
```

```python
# контейнер нормализующего потока: слои + априор, лог-правдоподобие и порождение. [Семинар 09: 09_Flows.ipynb]
class NormalizingFlow(nn.Module):

    def __init__(self, layers, prior):
        super(NormalizingFlow, self).__init__()
        self.layers = nn.ModuleList(layers)
        self.prior = prior

    def log_prob(self, x):
        # проход по слоям f, накопление суммы лог-определителей Якобиана.
        log_likelihood = None
        for layer in self.layers:
            x, log_det = layer.f(x)
            log_likelihood = log_det if log_likelihood is None else log_likelihood + log_det
        log_likelihood = log_likelihood + self.prior.log_prob(x)
        return log_likelihood.mean()

    def sample(self, n):
        z = self.prior.sample((n,))
        for layer in reversed(self.layers):
            z = layer.g(z)
        return z

# итог ячейки:
# - Что сделала: реализовала контейнер NormalizingFlow (лог-правдоподобие и порождение).
# - Что получилось: nf.log_prob(x) считает средний лог-правдоподобие, nf.sample(n) порождает выборку. источник: [Семинар 09: 09_Flows.ipynb]
```

### обучение

функция train_nf и обучение потока.

```python
import torch.nn as nn

class RealNVP(nn.Module):
    def __init__(self, var_size, mask, hidden=100):
        super(RealNVP, self).__init__()
        self.mask = mask  # вам это может не понадобиться. Восприниматйте как подсказку
        self.var_size = var_size

        # прямая и обратная сети t и s (зависимость от замаскинной части x). [Семинар 09: 09_Flows.ipynb]
        self.nn_t = nn.Sequential(
            nn.Linear(var_size, hidden),
            nn.ReLU(),
            nn.Linear(hidden, var_size),
        )
        self.nn_s = nn.Sequential(
            nn.Linear(var_size, hidden),
            nn.ReLU(),
            nn.Linear(hidden, var_size),
        )

    def f(self, x):
        # маска на устройстве и форме батча. [Семинар 09: 09_Flows.ipynb]
        mask = self.mask if isinstance(self.mask, torch.Tensor) else torch.tensor(self.mask, dtype=x.dtype, device=x.device)
        if mask.dim() == 1:
            mask = mask.unsqueeze(0).expand(x.shape[0], -1)
        x1 = x * mask
        t = self.nn_t(x1) * (1 - mask)
        s = self.nn_s(x1) * (1 - mask)
        s = torch.clamp(s, -5, 5)  # s ограничиваю, чтобы exp(s) не улетал в бесконечность
        y = x1 + (1 - mask) * (x * torch.exp(s) + t)
        log_det = (s * (1 - mask)).sum(dim=1)
        return y, log_det

    def g(self, y):
        mask = self.mask if isinstance(self.mask, torch.Tensor) else torch.tensor(self.mask, dtype=y.dtype, device=y.device)
        if mask.dim() == 1:
            mask = mask.unsqueeze(0).expand(y.shape[0], -1)
        y1 = y * mask
        t = self.nn_t(y1) * (1 - mask)
        s = self.nn_s(y1) * (1 - mask)
        s = torch.clamp(s, -5, 5)
        x = y1 + (1 - mask) * (y - t) * torch.exp(-s)
        return x
# итог ячейки:
# - что сделала: реализовала слой RealNVP. в слоях realnvp по очереди меняю, какие признаки меняются, а какие просто проходят дальше. источник: [Семинар 09: 09_Flows.ipynb]
```

```python
def train_nf(nf, X_train, n_epochs, batch_size, lr, device):
    # загрузчик и оптимизатор. [Семинар 09: 09_Flows.ipynb]
    X_t = torch.tensor(X_train, dtype=torch.float32, device=device)
    loader = torch.utils.data.DataLoader(
        torch.utils.data.TensorDataset(X_t),
        batch_size=batch_size,
        shuffle=True,
    )
    opt = torch.optim.Adam(nf.parameters(), lr=lr)
    nf.train()
    loss_history = []
    for epoch in range(n_epochs):
        for batch in loader:
            x = batch[0]
            opt.zero_grad()
            loss = -nf.log_prob(x)
            loss.backward()
            opt.step()
            loss_history.append(loss.item())
    return loss_history

# итог ячейки:
# - Что сделала: функция обучения потока по -log_prob, создаёт DataLoader и Adam внутри.
# - Что получилось: loss_history для графика обучения. источник: [Семинар 09: 09_Flows.ipynb]
```

```python
# вспомогательная функция: ROC-AUC по fake vs real в исходной шкале. [Семинар 08: 08_GANs.ipynb]
def eval_roc_from_fake(X_fake_train_orig, X_fake_test_orig):
    XX_train = np.concatenate((X_fake_train_orig, X_train_orig), axis=0)
    XX_test = np.concatenate((X_fake_test_orig, X_test_orig), axis=0)
    yy_train = np.array([0] * len(X_fake_train_orig) + [1] * len(X_train_orig))
    yy_test = np.array([0] * len(X_fake_test_orig) + [1] * len(X_test_orig))
    clf = GradientBoostingClassifier()
    clf.fit(XX_train, yy_train)
    yy_test_proba = clf.predict_proba(XX_test)[:, 1]
    return roc_auc_score(yy_test, yy_test_proba)

# мини-сетки параметров (короткие прогоны)
grid_wgan = [{"latent_dim": ld, "lr": lr} for ld in [8, 16, 32] for lr in [1e-4, 5e-5]]
grid_cvae = [{"latent_dim": ld, "lr": lr} for ld in [8, 16, 32] for lr in [1e-3, 5e-4]]
grid_diff = [{"hidden": h, "lr": lr, "infer_steps": s} for h in [64, 128] for lr in [1e-3, 5e-4] for s in [50, 100]]
grid_flows = [{"n_layers": n, "hidden": h} for n in [4, 6] for h in [64, 128]]

need = ["qt", "X_train_orig", "X_test_orig", "X_train", "X_test", "y_train", "y_test"]
missing = [v for v in need if v not in globals()]
if missing:
    raise NameError(f"Не найдены переменные: {missing}. Запусти блок подготовки данных.")

results = []
if not RUN_HEAVY:
    print("Пропуск мини-тюнинга. Запускать в Colab с RUN_HEAVY=True.")
else:
    import time
    d = X_train.shape[1]
    # WGAN
    for p in grid_wgan:
        t0 = time.perf_counter()
        gen = Generator(n_inputs=p["latent_dim"] + y_train.shape[1], n_outputs=d)
        disc = Discriminator(n_inputs=d + y_train.shape[1])
        fitter = Fitter(gen, disc, batch_size=TUNE_BATCH, n_epochs=TUNE_EPOCHS_WGAN, latent_dim=p["latent_dim"], lr=p["lr"], n_critic=3)
        fitter.fit(X_train, y_train)
        X_fake_tr = generate(fitter.generator, y_train, p["latent_dim"])
        X_fake_te = generate(fitter.generator, y_test, p["latent_dim"])
        X_fake_tr_orig = qt.inverse_transform(X_fake_tr)
        X_fake_te_orig = qt.inverse_transform(X_fake_te)
        auc = eval_roc_from_fake(X_fake_tr_orig, X_fake_te_orig)
        results.append({"model": "wgan", **p, "auc": float(auc), "time_sec": round(time.perf_counter() - t0, 1), "note": "тестовый прогон"})
    # cVAE
    for p in grid_cvae:
        t0 = time.perf_counter()
        enc = Encoder(n_inputs=d + y_train.shape[1], lat_size=p["latent_dim"])
        dec = Decoder(n_inputs=p["latent_dim"] + y_train.shape[1], n_outputs=d)
        vae = VAEFitter(enc, dec, batch_size=TUNE_BATCH, n_epochs=TUNE_EPOCHS_CVAE, latent_dim=p["latent_dim"], lr=p["lr"], KL_weight=0.001)
        vae.fit(X_train, y_train)
        X_fake_tr = generate(vae.decoder, y_train, p["latent_dim"])
        X_fake_te = generate(vae.decoder, y_test, p["latent_dim"])
        X_fake_tr_orig = qt.inverse_transform(X_fake_tr)
        X_fake_te_orig = qt.inverse_transform(X_fake_te)
        auc = eval_roc_from_fake(X_fake_tr_orig, X_fake_te_orig)
        results.append({"model": "cvae", **p, "auc": float(auc), "time_sec": round(time.perf_counter() - t0, 1), "note": "тестовый прогон"})
    # Diffusion
    for p in grid_diff:
        t0 = time.perf_counter()
        model = DiffusionGenerator(X_train.shape[1] + y_train.shape[1], d, time_embedding_dim=p["hidden"]).to(DEVICE)
        dfitter = DiffusionFitter(model, batch_size=TUNE_BATCH, n_epochs=TUNE_EPOCHS_DIFF, latent_dim=10, lr=p["lr"])
        dfitter.fit(X_train, y_train)
        sched = DDPMScheduler(num_train_timesteps=1000)
        X_fake_tr = generate_with_diffusion(dfitter.model, y_train, sched, num_inference_steps=p["infer_steps"])
        X_fake_te = generate_with_diffusion(dfitter.model, y_test, sched, num_inference_steps=p["infer_steps"])
        X_fake_tr_orig = qt.inverse_transform(X_fake_tr)
        X_fake_te_orig = qt.inverse_transform(X_fake_te)
        auc = eval_roc_from_fake(X_fake_tr_orig, X_fake_te_orig)
        results.append({"model": "diffusion", **p, "auc": float(auc), "time_sec": round(time.perf_counter() - t0, 1), "note": "тестовый прогон"})
    # Flows
    for p in grid_flows:
        t0 = time.perf_counter()
        prior = torch.distributions.MultivariateNormal(torch.zeros(d, device=DEVICE), covariance_matrix=torch.eye(d, device=DEVICE))
        mask0 = (torch.arange(d, device=DEVICE) % 2).float()
        mask1 = 1 - mask0
        layers = []
        for i in range(p["n_layers"]):
            mask = mask0 if i % 2 == 0 else mask1
            layers.append(RealNVP(var_size=d, mask=mask, hidden=p["hidden"]))
        nf = NormalizingFlow(layers=layers, prior=prior).to(DEVICE)
        train_nf(nf, X_train, TUNE_EPOCHS_FLOWS, TUNE_BATCH, NF_LR, DEVICE)
        nf.eval()
        with torch.no_grad():
            X_fake_tr = nf.sample(len(X_train)).cpu().numpy()
            X_fake_te = nf.sample(len(X_test)).cpu().numpy()
        X_fake_tr_orig = qt.inverse_transform(X_fake_tr)
        X_fake_te_orig = qt.inverse_transform(X_fake_te)
        auc = eval_roc_from_fake(X_fake_tr_orig, X_fake_te_orig)
        results.append({"model": "flows", **p, "auc": float(auc), "time_sec": round(time.perf_counter() - t0, 1), "note": "тестовый прогон"})

if results:
    df = pd.DataFrame(results).sort_values(["model", "auc"], ascending=[True, True])
    print("Лучшие 20 строк (auc ближе к 0.5 - лучше):")
    print(df.head(20).to_string())
    (ART_DIR / "tuning").mkdir(parents=True, exist_ok=True)
    df.to_csv(ART_DIR / "tuning" / "tuning_results.csv", index=False)
    safe_save("tuning", extra={"results": results})
    for mod in ["wgan", "cvae", "diffusion", "flows"]:
        sub = df[df["model"] == mod]
        if len(sub) > 0:
            best = sub.loc[sub["auc"].idxmin()]
            params = {k: best[k] for k in ["latent_dim", "lr", "hidden", "n_layers", "infer_steps"] if k in best.index}
            print(f"Лучший набор для {mod}: auc={best['auc']:.4f} (ближе к 0.5 - лучше), params={params}")
    print("Это быстрый подбор параметров на малом числе эпох, нужен чтобы сравнить направления, финальные значения можно уточнять при большем времени.")

# итог ячейки:
# - Что сделала: добавила быстрый мини-подбор параметров для всех моделей, с короткими прогонами и сохранением результатов.
# - Что получилось: таблица результатов и файл artifacts_hw3/tuning/results.json (и tuning_results.csv). источник: [Семинар 08: 08_GANs.ipynb], [Семинар 08: 08_Autoencoders.ipynb], [Семинар 09: 09_Diffusions.ipynb], [Семинар 09: 09_Flows.ipynb]
# - Критерий выбора - минимальный ROC-AUC (ближе к 0.5 - лучше).
```

качество проверяли метрикой roc auc. идея простая:  смешиваем настоящие и сгенерированные объекты и учим классификатор их различать. если он не может отличить фейк от настоящего значит генерация хорошая, и roc auc будет около 0.5. чем выше roc auc, тем легче классификатору отличить подмену, то есть тем хуже генерация.
